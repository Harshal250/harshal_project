```python
import pandas as pd

# Reading the CSV file into a DataFrame
df1 = pd.read_csv('imdb.title.basics (1).csv')

# Splitting the 'genres' column into a list
df1['genres'] = df1['genres'].str.split(',')

# Exploding the 'genres' column
df1_exploded = df1.explode('genres')

# Displaying the first few rows of the DataFrame with exploded 'genres'
df1_exploded.head()
```

```python
# Reading the CSV file into a DataFrame
df2 = pd.read_csv('tn.movie_budgets (1).csv')

# Converting 'production_budget', 'worldwide_gross', and 'domestic_gross' columns to numeric
df2['production_budget'] = df2['production_budget'].replace('[\$,]', '', regex=True).astype(float)
df2['worldwide_gross'] = df2['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)
df2['domestic_gross'] = df2['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)

# Calculating profit as the difference between worldwide_gross and production_budget
df2['profit'] = df2['worldwide_gross'] - df2['production_budget']

# Converting values in production_budget, worldwide_gross, and profit into million
df2['production_budget'] /= 1000000
df2['worldwide_gross'] /= 1000000
df2['domestic_gross'] /= 1000000
df2['profit'] /= 1000000

# Adding dollar symbol to all values
df2['production_budget'] = '$' + df2['production_budget'].astype(str)
df2['worldwide_gross'] = '$' + df2['worldwide_gross'].astype(str)
df2['domestic_gross'] = '$' + df2['domestic_gross'].astype(str)
df2['profit'] = '$' + df2['profit'].astype(str)

# Displaying the first few rows of the DataFrame with values in million and dollar symbols
df2.head()
```

```python
# Merging df1_exploded and df2 on the 'title' and 'genres' columns with an outer join
merged_df = pd.merge(df1_exploded, df2, on=['title'], how='right')

# Displaying the first few rows of the merged DataFrame
merged_df.head()
```

```python
# Importing the pandas library
import pandas as pd

# Reading the CSV file into a DataFrame
df3 = pd.read_csv('imdb.title.ratings.csv.gz')

# Displaying the first few rows of the DataFrame
df3.head()
```

```python
# Merging merged_df and df3 on the 'tconst' column with a right join
final_merged_df = pd.merge(merged_df, df3, on='tconst', how='left')

# Displaying the first few rows of the final merged DataFrame
final_merged_df.head(100)
```

```python
import pandas as pd

# Assuming final_merged_df is the merged DataFrame
# Replace these column names with the actual columns in your DataFrame

# Converting 'production_budget' and 'worldwide_gross' columns to numeric
final_merged_df['production_budget'] = final_merged_df['production_budget'].replace('[\$,]', '', regex=True).astype(float)
final_merged_df['worldwide_gross'] = final_merged_df['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)

# Calculating profit as the difference between worldwide_gross and production_budget
final_merged_df['profit'] = final_merged_df['worldwide_gross'] - final_merged_df['production_budget']

# Calculating Return on Investment (ROI)
final_merged_df['roi'] = (final_merged_df['profit'] / final_merged_df['production_budget']) * 100

# Displaying the first few rows of the DataFrame with the new 'roi' column
display(final_merged_df.head(100))




```

```python
import pandas as pd

# Assuming final_merged_df is your DataFrame
# Replace 'roi' with the actual column name in your DataFrame

# Convert 'roi' column to numeric, coerce non-numeric values to NaN
final_merged_df['roi'] = pd.to_numeric(final_merged_df['roi'], errors='coerce')

# Check for any NaN values in 'roi' column
non_numeric_roi = final_merged_df['roi'].isna().any()

if non_numeric_roi:
    print("There are non-numeric values in the 'roi' column.")
else:
    print("All values in the 'roi' column are numeric.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming final_merged_df is the DataFrame with 'genres' and 'profit' columns
# Replace these column names with the actual columns in your DataFrame

# Sorting DataFrame by profit in descending order
top_genres_df = final_merged_df.sort_values(by='profit', ascending=False).head(6)

# Convert 'genres' to strings
top_genres_df['genres'] = top_genres_df['genres'].astype(str)

# Creating a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_genres_df['genres'], top_genres_df['profit'], color='skyblue')
plt.title('Top 5 Highest Profitable Genres')
plt.xlabel('Genres')
plt.ylabel('Total Profit (in million dollars)')
plt.xticks(rotation=45, ha='right')
plt.show()


```

```python
# Extracting 'genres' and 'roi' columns
genres_roi_data = final_merged_df[['genres', 'roi']]

# Displaying the extracted data
display(genres_roi_data)
```

```python
# Removing NaN values from 'genres' and 'roi' columns
genres_roi_data_cleaned = genres_roi_data.dropna()

# Displaying the cleaned data
display(genres_roi_data_cleaned)
```

```python
import matplotlib.pyplot as plt

# Sorting the DataFrame by 'roi' in descending order and selecting the top 5 rows
top_genres_roi_df = genres_roi_data_cleaned.sort_values(by='roi', ascending=False).head(5)

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_genres_roi_df['genres'], top_genres_roi_df['roi'], color='skyblue')
plt.title('Top 5 Genres with Highest ROI')
plt.xlabel('Genres')
plt.ylabel('ROI (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Importing the pandas library
import pandas as pd

# Reading the CSV file into a DataFrame
df5 = pd.read_csv('bom.movie_gross.csv.gz')

# Displaying the first few rows of the DataFrame
df5.head()
```

```python
# Merge df5 and final_merged_df on the 'title' column
merged_df = pd.merge(df5, final_merged_df, on='title', how='inner')

# Display the first few rows of the merged DataFrame
merged_df.head(10)
```

```python
print(merged_df.columns)
```

```python
import matplotlib.pyplot as plt

# Grouping by 'studio' and calculating the mean of 'roi' for each studio
studio_roi_means = merged_df.groupby('studio')['roi'].mean()

# Sorting the studios based on mean ROI in descending order
top_studios = studio_roi_means.sort_values(ascending=False).head(5)

# Creating a bar chart
plt.figure(figsize=(12, 6))
top_studios.plot(kind='bar', color='skyblue')
plt.title('Top 5 Studios with Highest ROI')
plt.xlabel('Studio')
plt.ylabel('Average ROI')
plt.show()
```
